# Use gradient descent to find a minimum for cost function 4x^4-3x^2+2x-1

```python
import numpy as np
import matplotlib.pyplot as plt
```

Gradient descent requires the derivative of the cost functon. This is given by:
![title](cost function.PNG)

```python
# Define f(x)
def f(x):
    return 4*(x**4) - 3*(x**2) + 2*x -1

# Define the derivative of f(x)
def f_d(x):
    return 16*x**3 - 6*x +2
```

```python
# Perform Gradient Descent
funcLimits=[-2, 1, -4, 1]

oldX = np.random.random()*(funcLimits[1]-funcLimits[0]) + funcLimits[0]
YHistory = [f(oldX)]

deltaX=oldX
for epoch in range(0, maxEpochs):
    plotFunc(f, oldX, funcLimits, "Epoch="+str(epoch))
    
    newX = oldX - alphaDecay(epoch)*f_d(oldX)
    deltaX = newX-oldX
    
    if (deltaX == 0):
        break;

    print newX
    YHistory = YHistory + [f(newX)]
    oldX = newX
    
print("Minimum=",newX)
print("Epoch=", epoch)
print("DeltaX=", deltaX)
print("YHistory=", YHistory)
```
